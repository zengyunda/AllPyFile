```python
import urllib.request
import re,xlwt
from bs4 import BeautifulSoup
list11=[]
list21=[]
list31=[]
list41=[]
list51=[]
list61=[]
for page in range(0,276,25):
    html=urllib.request.urlopen('https://www.douban.com/doulist/240962/?start=75').read()
    list1=re.compile(r'<div class="title">.*? target="_blank">(.*?)</a>',re.S).findall(html.decode())
    list2=re.compile(r'<span class="allstar(.*?)"></span>',re.S).findall(html.decode())
    list3=re.compile(r'<span class="rating_nums">(.*?)</span>',re.S).findall(html.decode())
    list4=re.compile(r'<div class="abstract">(.*?) </div>',re.S).findall(html.decode())
    list5=re.compile(r'<blockquote class="comment">(.*?) </blockquote>',re.S).findall(html.decode())
    list6=re.compile(r'a href="(https://movie.douban.com/.*?)" target="_blank">',re.S).findall(html.decode())
    list11.extend(list1)
    list21.extend(list2)
    list31.extend(list3)
    list41.extend(list4)
    list51.extend(list5)
    list61.extend(list6)
workbook=xlwt.Workbook()
sheet=workbook.add_sheet('douban')
j=0
list11.insert(0,'电影名称')
list21.insert(0,'星级')
list31.insert(0,'评分')
list41.insert(0,'导演及演员信息')
list51.insert(0,'评价')
list61.insert(0,'链接网址')
for i in range(0,len(list11)):
    sheet.write(i,j,list11[i])
    sheet.write(i,j+1,list21[i])
    sheet.write(i,j+2,list31[i])
    sheet.write(i,j+3,list41[i])
    sheet.write(i,j+4,list51[i])
    sheet.write(i,j+5,list61[i])           
workbook.save('D:/豆瓣高分电影.xls')
```
