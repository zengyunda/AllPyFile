```python
import requests
from pyquery import PyQuery as pq
import pandas as pd
import matplotlib.pyplot as plt
```

```python
list_all=[]
for pages in range(1,101):
    res = requests.get('http://esf.qingyuan.fang.com/house-a012885-b015044/i3'+str(pages))
    res = res.content.decode('gbk')
    doc = pq(res)  
    print('第'+str(pages)+'页')
    for i in doc('.houseList p').items():
        list_all.append(i.text())
print(len(list_all))
print(list_all)
```

```python
for i in list_all[1::8]:
    print(i)
    #print(i.split('|')[3].split(':'))
```

```python
pf = pd.DataFrame({
    'title':list_all[::8],
    'age':list_all[1::8],
    'cell_neme':list_all[2::8],
    'agent':list_all[3::8],
    'area':list_all[4::8],
    'price_total':list_all[6::8],
    'price_single':list_all[7::8]
})
print(pf.head(5))
```

```python
pf['size'], pf['hige'] = zip(*pf['age'].map(lambda x: x.split('|')))

pf.head(5)
```

```python
pf.pop('cell_neme')
pf.head(5)
```

```python
pz = pf[['size','age']].groupby(['size']).count().sort_values(by='age',ascending=False).reset_index()
plt.hist(pz['age'],30,normed=True)
plt.show
```
